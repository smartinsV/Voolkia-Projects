```python
import pandas as pd
from pathlib import Path
```

```python
file_note = Path("output_dataset") / "merged_transformed.feather"
file_api = Path("output_dataset") / "merged_transformed_api.feather"
```

```python
df_note = pd.read_feather(file_note)
df_note.head()
```

```python
note_nulls = (df_note == -1).sum()
note_nulls
```

```python
df_api = pd.read_feather(file_api)
df_api.head()
```

```python
indx = df_api[df_api["NUM_SECU_EXPED"].isin(set(df_note["NUM_SECU_EXPED"]))].index
df_api = df_api.loc[indx]
```

```python
api_nulls = (df_api == -1).sum()
api_nulls
```

```python
full_cols = set(api_nulls.index).intersection(set(note_nulls.index))
for c in full_cols:
    print(c)
    print(df_api[c].value_counts())
    print(df_note[c].value_counts())
    print("*/-*/-")
```

```python
df_api = df_api[full_cols].reset_index(drop=True)
```

```python
df_note[full_cols].equals(df_api[full_cols])
```

```python
df_note[full_cols].head()
```

```python
df_api[full_cols].head()
```

```python
for i in range(df_api.shape[0]):
    for c in full_cols:
        print(df_api.loc[i, c], df_note.loc[i, c], df_api.loc[i, c]!= df_note.loc[i, c])
        if df_api.loc[i, c]!= df_note.loc[i, c]:
            print(i, c)
```
