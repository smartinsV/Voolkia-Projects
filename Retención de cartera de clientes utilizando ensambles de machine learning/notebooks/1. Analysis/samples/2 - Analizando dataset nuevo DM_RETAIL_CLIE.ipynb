## DM_RETAIL_CLIE

```python
chunks = pd.read_csv(f"../Datasets/DM_RETAIL_CLIE.csv", 
                     sep=",",
                     encoding='latin1',
                     decimal=",",
                     chunksize=1000000)

df_retail = pd.concat(chunks)
```

```python
df_retail.isna().mean().sort_values(ascending=False)
```

```python
df_retail = df_retail.loc[:,df_retail.isna().mean() < .8]
```

```python
fechas = [col for col in df_retail.columns if "F_U" in col or "F_C" in col or "FECHA" in col]
df_retail[fechas]
```

```python
df_retail = df_retail.drop(fechas, axis=1)
```
