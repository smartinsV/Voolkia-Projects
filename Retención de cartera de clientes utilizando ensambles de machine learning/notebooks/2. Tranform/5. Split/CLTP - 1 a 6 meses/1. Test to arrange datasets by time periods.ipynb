## Armado de dataset para entrenamiento del modelo

Se usan ventanas de tiempo de 6 meses para agrupar los datos y predecir la baja

```python
df = pd.read_feather('../../../data/processed/merged_data_18_periods.feather')
```

```python
df
```

```python
import re
```

```python
periodic_cols = set([''.join([s for s in col if not s.isdigit()]) for col in df[[col for col in df.columns if re.search(f'\d', col)]].columns])
```

```python
periodic_cols
```

```python
df[[col for col in df.columns if re.search(f'_[1-6]$|^[1-6]_', col)]].columns
```

**Prueba para primer recorte** 
- Periodos 1 a 6

**Removemos todas las columnas temporales y las guardamos aparte**

```python
grouped_df = df[[col for col in df.columns if not any(char.isdigit() for char in col)]]
```

```python
grouped_df
```

**Agregamos sumatorias de las columnas temporales**

```python
for pcol in periodic_cols:
    columns = [col for col in df.columns if re.search(f'{pcol}[1-6]$|^[1-6]{pcol}', col)]
    grouped_df[pcol] = df[columns].sum(axis=1)
```

**Eliminamos las filas que no tienen polizas activas en el periodo o ya se dieron de baja**

Buscamos los que no tienen polizas en el periodo analizado

```python
grouped_df[grouped_df["hist_polizas"].str.split(" ", expand=True).iloc[:,0:5].astype(float).sum(axis=1) == 0]
```

```python
grouped_df = grouped_df[grouped_df["hist_polizas"].str.split(" ", expand=True).iloc[:,0:5].astype(float).sum(axis=1) > 0]
```

Buscamos las polizas que ya se dieron de baja

```python
grouped_df[grouped_df["periodo_baja"].between(1,6)]
```

```python
grouped_df = grouped_df[~grouped_df["periodo_baja"].between(1,6)]
```

**Marcamos aquellos que se dieron de baja en los tres meses proximos la periodo analizado**

```python
(grouped_df["periodo_baja"] > 0).value_counts(normalize=True)
```

```python
grouped_df["BAJA_3m"] = grouped_df["periodo_baja"].between(7,9)
```

```python
grouped_df = grouped_df.drop(["CIF_ID","hist_polizas","periodo_baja"], axis=1)
```
