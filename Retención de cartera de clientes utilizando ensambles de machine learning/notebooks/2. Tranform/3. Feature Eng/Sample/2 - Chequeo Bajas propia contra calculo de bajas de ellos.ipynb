## Observaciones para las personas con el campo cliente_perdido

```python
df_polizas = pd.read_feather('../../../data/interim/polizas.feather')
```

```python
df_polizas['mes_anio_vig'] = df_polizas['FECHA_VIG_POL'].dt.strftime('%Y-%m')
```

```python
df_polizas_min = df_polizas[["CIF_ID","NUM_SECU_POL",'mes_anio_vig','MCA_VIGENCIA']].drop_duplicates()
```

```python
df_polizas_pivoted = df_polizas_min.pivot_table(index='CIF_ID',columns=['mes_anio_vig'], values=['MCA_VIGENCIA'],aggfunc='count',fill_value=0)
```
