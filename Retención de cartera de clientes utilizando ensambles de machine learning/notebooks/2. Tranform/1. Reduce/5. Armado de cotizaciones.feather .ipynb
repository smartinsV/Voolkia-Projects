## Creating cotizaciones.feahter

```python
chunks = pd.read_csv(f"{location}/{table}",
                     sep=sep,
                     encoding=encoding,
                     decimal=decimal,
                     chunksize=1000000,
                     iterator=True,
                     usecols = usable_cols)
df = pd.concat(chunks)
```

```python
objects = [col for col in df.columns if df[col].dtype =="object"]
```

```python
for obj in objects: 
    print(f"{obj}:{df[obj].unique()}")
    print("---------------------------")
```
