## Creating interacciones.feahter

```python
chunks = pd.read_csv(f"{location}/{table}",
                     sep=sep,
                     encoding=encoding,
                     decimal=decimal,
                     chunksize=1000000,
                     iterator=True,
                     usecols = usable_cols)
df = pd.concat(chunks)
```

```python
df[pd.to_numeric(df['ID'], errors='coerce').isnull()]
```

*Cleaning weird values in ID and CIF_ID* 

```python
df = df[~pd.to_numeric(df['ID'], errors='coerce').isnull()]
```

```python
to_check = []
for val in df["CIF_ID"].unique():
    try:
        float(val)
    except Exception as e:
        print(e)
        to_check.append(val)
```

```python
df[df["CIF_ID"].isin(to_check)]
```

```python
df = df[~df["CIF_ID"].isin(to_check)]
```

```python
df["CIF_ID"].isna().sum()
```

```python
df["ID"].isna().sum()
```

```python
to_check = []
for val in df["ID"].unique():
    try:
        int(val)
    except Exception as e:
        print(e)
        to_check.append(val)
```

```python
df[df["ID"].isin(to_check)]
```

```python
df = df[~df["ID"].isin(to_check)]
```

```python
df = df.drop(columns='ID').astype({'CIF_ID': 'float64'})
```

```python
df.reset_index(drop=True).to_feather("../../../data/interim/interacciones.feather")
```
