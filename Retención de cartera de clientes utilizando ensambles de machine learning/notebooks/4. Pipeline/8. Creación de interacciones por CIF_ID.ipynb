```python
df["FECHA"] = df["FECHA"].str.slice(stop=10) 
```

```python
df.loc[df["FECHA"].str.contains(" [0-9]", na=False), "FECHA"] = df.loc[df["FECHA"].str.contains(" [0-9]", na=False), "FECHA"].str.slice(stop=8) 
```

```python
df["FECHA"] = df["FECHA"].str.replace(" ","")
```

```python
def lookup(s):
    """
    This is an extremely fast approach to datetime parsing.
    For large data, the same dates are often repeated. Rather than
    re-parse these, we store all unique dates, parse them, and
    use a lookup to convert all dates.
    """
    dates = {date:pd.to_datetime(date, format="%d/%m/%Y") for date in s.unique()}
    return s.map(dates)

df["FECHA"] = lookup(df["FECHA"])
```

Transformamos la fecha en anio y mes

```python
df['mes_anio_interacc'] = df["FECHA"].dt.strftime('%Y-%m')
```

```python
df = df[["CIF_ID","IN_OUT","CANAL_COMUNICACION","GESTION","DESCRIPCION","AGRUPACION","mes_anio_interacc"]]
```

```python
df.rename(columns={"mes_anio_interacc": "periodo"}, inplace=True)
```
