```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/processed/interm-BD_INT_PAGO_POLIZA_demora_pago.completo.csv", index_col="Unnamed: 0")
df.head()
```

```python
columns = df.corresp_date.unique()
row = df.CUSTOMER_ID.unique()
```

```python
#creo el dataset vacio
df_m = pd.DataFrame(columns=columns)
df_m.info()
```

```python
#agrupo por cliente y mes_anio y cuento cuantas polizas distintas tiene
groups = df.groupby(["CUSTOMER_ID", "corresp_date"])["POLICY_NUMBER"].count()
```

```python
for id_cliente in groups.index.levels[0]:
    df_m.loc[id_cliente] = groups[id_cliente].T
```

```python
df_m.head()
```

```python
df_m.to_csv("../../data/processed/BD_ID_CLIENTE_N.PAGO.POLIZA_X_MES.csv")
```
