```python
df_dif = df - df_canc
df_dif = df_dif == 0
df_dif.head()
```

# armando bloques 9 m (6 train, 3 target)

### Creo lista de dataframes

```python
#descarto las últimas 2 cols porque las uso para testeo (interno y cliente)
df = df.iloc[:,:-2]
df_canc = df_canc.iloc[:,:-2]
df_dem = df_dem.iloc[:,1:-2]
df_deu = df_deu.iloc[:,1:-2]
df_dif = df_dif.iloc[:,1:-2]
df.info()
```

```python
df_deu.info()
```

```python
len(df.columns)
```

```python
#termino en 5 para que no de error en el rango de las columnas
list_df = list()
list_df_canc = list()
list_df_dem = list()
list_df_deu = list()
list_df_dif = list()
for i in range( (len(df.columns)) ,8, -1):
    #print("{}-{}".format(i-9,i))
    #print("{}-{}".format(i-9,i-3))
    list_df.append(df.iloc[:,i-9:i])
    list_df_canc.append(df_canc.iloc[:,i-9:i-3])
    list_df_dem.append(df_dem.iloc[:,i-9:i-3])
    list_df_deu.append(df_deu.iloc[:,i-9:i-3])
    list_df_dif.append(df_dif.iloc[:,i-9:i-3])

len(list_df)
```

```python
tmp = list()
for i in range(0, len(list_df)):
    tmp.append(pd.merge(list_df[i], list_df_canc[i], how="left", left_index=True, right_index=True, suffixes=["","_canc"]))
    tmp[i] = pd.merge(tmp[i], list_df_dem[i], left_index=True, right_index=True, suffixes=["","_dem"])
    tmp[i] = pd.merge(tmp[i], list_df_deu[i], left_index=True, right_index=True, suffixes=["","_deu"])
    tmp[i] = pd.merge(tmp[i], list_df_dif[i], left_index=True, right_index=True, suffixes=["","_did"])
len(tmp)
```

```python
#chequeando integridad de valores
#tmp[0].loc[7968,:]
```

### Renombre las columnas de cada dataframe

```python
#tmp[0].columns
cols = ["6m", "5m", "4m", "3m", "2m", "1m", "3p", "2p", "1p",
        '6m_canc', '5m_canc', '4m_canc', '3m_canc', '2m_canc', '1m_canc', 
       '6m_dem', '5m_dem', '4m_dem', '3m_dem', '2m_dem', '1m_dem',
       '6m_deu', '5m_deu', '4m_deu', '3m_deu', '2m_deu', '1m_deu',
       '6m_dif', '5m_dif', '4m_dif', '3m_dif', '2m_dif', '1m_dif']
```

```python
#cambiando el nombre de las fechas a versión generica ( 6 meses train, 3 pred)
for i in range(0, len(tmp)):
    tmp[i].rename(columns=dict(zip(tmp[i].columns, cols)),inplace=True)
```

### Integrando lista de dataframes en un único dataframe

```python
bd = pd.concat(tmp)
bd.info()
```

### Integrando con dataframe clientes

```python
df_int = pd.merge(bd, df_cli, left_index=True, right_index=True)
df_int.info()
```

```python
df_int["ID_CUSTOMER"] = df_int.index
df_int.head()
```

```python
df_int.drop_duplicates(inplace=True)
df_int.info()
```

```python
df_int.drop("ID_CUSTOMER", axis=1, inplace=True)
df_int.head()
```

```python
#variables descartadas
delete_cols = ["DATE_OF_BIRTH", "FECHA_PROCESO", "OCCUPATION", "LOCALITY", "edad"]
df_fin = df_int.drop(delete_cols, axis=1)
df_fin.info()
```

```python
X = pd.get_dummies(df_fin)
X.info()
```

```python
X.fillna(0, inplace=True) #reemplazo nulos por 0
X = X.astype("uint8") #cambio el tipo de los datos
X.loc[:,['6m_dif', '5m_dif', '4m_dif', '3m_dif', '2m_dif', '1m_dif']] = X.loc[:,['6m_dif', '5m_dif', '4m_dif', '3m_dif', '2m_dif', '1m_dif']].astype("bool")
X.info()
```

```python
X["id_cliente"] = X.index
```

```python
X.drop_duplicates(inplace=True)
X.info()
```
