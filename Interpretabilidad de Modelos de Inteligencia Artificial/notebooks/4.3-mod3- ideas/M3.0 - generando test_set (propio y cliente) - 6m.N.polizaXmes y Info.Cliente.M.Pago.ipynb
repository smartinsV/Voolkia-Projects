```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
#leer la bd
df = pd.read_csv("../../data/processed/BD_ID_CLIENTE_N.POLIZA_X_MES.csv", index_col="Unnamed: 0")
df.head()
```

```python
#agrego la bd de datos cliente
df_cli = pd.read_csv("../../data/processed/BD_ID_CLIENTE_D.Pers_Modo.Pago.csv", index_col="Unnamed: 0")
df_cli.set_index("CUSTOMER_ID", inplace=True)
df_cli.sort_index(inplace=True)
df_cli.head()
```

```python
df_canc = pd.read_csv("../../data/processed/BD_ID_CLIENTE_N.CANCELACIONES.negativas_X_MES.csv", index_col="Unnamed: 0")
df_canc["2018-01"] = np.nan
df_canc.head()
```

```python
#matriz demora pago
df_dem = pd.read_csv("../../data/processed/BD_ID_CLIENTE_N.DEMORA_X_MES.csv", index_col="Unnamed: 0")
df_dem.head()
```

```python
#matriz deuda pago
df_deu = pd.read_csv("../../data/processed/BD_ID_CLIENTE_N.pago.deuda_X_MES.csv", index_col="Unnamed: 0")
df_deu.head()
```

```python
df_dif = df - df_canc
df_dif = df_dif == 0
df_dif.head()
```

# Test set cliente

```python
#obtengo los últimos 6 meses - (3 siguientes a predecir)
bd = df.iloc[:,-6:]
bd.head()
```

```python
df_int = pd.merge(bd, df_cli, left_index=True, right_index=True)
df_int.info()
```

```python
#integrando los otros datasets: cancelacion
df_int = pd.merge(df_int, df_canc.iloc[:,-6:], how="left", left_index=True, right_index=True, suffixes=["","_canc"])
df_int.info()
```

```python
df_int.head()
```

```python
#integrando los otros datasets: cancelacion
df_int = pd.merge(df_int, df_dem.iloc[:,-6:], left_index=True, right_index=True, suffixes=["","_dem"])
df_int.info()
```

```python
#integrando los otros datasets: cancelacion
df_int = pd.merge(df_int, df_deu.iloc[:,-6:], left_index=True, right_index=True, suffixes=["","_deu"])
df_int.info()
```

```python
df_int.head()
```

```python
#integrando los otros datasets: cancelacion
df_int = pd.merge(df_int, df_dif.iloc[:,-6:], left_index=True, right_index=True, suffixes=["","_dif"])
df_int.head()
```

```python
#variables descartadas
delete_cols = ["DATE_OF_BIRTH", "FECHA_PROCESO", "OCCUPATION", "LOCALITY", "edad"]
df_fin = df_int.drop(delete_cols, axis=1)
df_fin.info()
```

```python
cols = ["6m", "5m", "4m", "3m", "2m", "1m",
       'GENDER', 'MARITAL_STATUS', 'EDUCATION', 'EDUCATION_LEVEL', 'CITY',
       'ModoPago', 'edad_r', 
       '6m_canc', '5m_canc', '4m_canc', '3m_canc', '2m_canc', '1m_canc', 
       '6m_dem', '5m_dem', '4m_dem', '3m_dem', '2m_dem', '1m_dem',
       '6m_deu', '5m_deu', '4m_deu', '3m_deu', '2m_deu', '1m_deu',
       '6m_dif', '5m_dif', '4m_dif', '3m_dif', '2m_dif', '1m_dif']
```

```python
#cambiando el nombre de las fechas a versión generica ( 6 meses model)
df_fin.rename(columns=dict(zip(df_fin.columns, cols)),inplace=True)
df_fin.info()
```

```python
X = pd.get_dummies(df_fin)
X.info()
```

```python
X.fillna(0, inplace=True) #reemplazo nulos por 0
X = X.astype("uint8") #cambio el tipo de los datos
X.loc[:,['6m_dif', '5m_dif', '4m_dif', '3m_dif', '2m_dif', '1m_dif']] = X.loc[:,['6m_dif', '5m_dif', '4m_dif', '3m_dif', '2m_dif', '1m_dif']].astype("bool")
X.info()
```

```python
X.head()
```

```python
X.to_csv("../../data/processed/test_cliente/m3.0 - 6m.N.polizaXmes y Info.Cliente.M.Pago.csv")
```
