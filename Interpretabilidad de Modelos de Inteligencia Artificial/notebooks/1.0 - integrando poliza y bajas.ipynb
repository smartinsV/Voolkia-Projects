## Controlemos si se repiten las polizas

```python
#df_poliza[df_poliza.groupby("POLICY_NUMBER").count()["CUSTOMER_ID"] > 1]
df_pol_group = pd.DataFrame(df_poliza.groupby("POLICY_NUMBER").count()["CUSTOMER_ID"] > 1)
df_pol_group.reset_index(level=0, inplace=True)
df_pol_group.info()
```

```python
(df_poliza.groupby("POLICY_NUMBER").count()["CUSTOMER_ID"] > 1).sum() / df_poliza["CUSTOMER_ID"].nunique()
```

```python
#obtengo todas las polizas que se repiten
#df_pol_group[df_pol_group["CUSTOMER_ID"] == True] #devuelve el subset de polizas que tienen más de 1 registro
ids = df_pol_group[df_pol_group["CUSTOMER_ID"] == True]["POLICY_NUMBER"].values
df_poliza[df_poliza.POLICY_NUMBER.isin(ids)].sort_values(by="POLICY_NUMBER")
```
