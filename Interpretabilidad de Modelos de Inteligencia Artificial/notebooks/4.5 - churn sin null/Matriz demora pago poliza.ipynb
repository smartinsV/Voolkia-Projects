Este script va a agregar a cada poliza el ID del cliente al que pertenece

```python
import pandas as pd
import numpy as np
```

```python
df_poliza = pd.read_table("../../data/raw/TB_POLIZAS_CL.tsv", encoding="latin-1")
df_poliza.info()
```

```python
df_poliza.head()
```

```python
df_poliza[df_poliza["POLICY_NUMBER"] == 77320269999]
```

```python
df_pago = pd.read_table("../../data/raw/TB_PAGOS_CL.tsv", encoding="latin-1", low_memory=False)
df_pago.info()
```

```python
df_pago[df_pago["POLICY_NUMBER"] == 253295029999]
```

```python
df_pag_pol = df_pago.merge(df_poliza[["POLICY_NUMBER", "CUSTOMER_ID"]], how="inner", left_on=["POLICY_NUMBER"], 
                             right_on=["POLICY_NUMBER"], suffixes=['', '_poliza'])
df_pag_pol.info()
```

```python
bd = df_pag_pol.drop_duplicates()
bd.info()
```

```python
bd[bd.POLICY_NUMBER.isnull()]
```

```python
bd.head()
```

```python
#convierto el due_date en fecha, para luego obtener el siguiente mes (correspondiente al servicio)
from datetime import datetime, timedelta
bd["fecha"] = bd.apply(lambda x : datetime.strptime("{}-{}-01".format(x["DUE_DATE"].split("-")[0], x["DUE_DATE"].split("-")[1]), "%Y-%m-%d"), axis=1)
bd.info()
```

```python
bd["next_f"] = bd.fecha.apply(lambda x : x if pd.isnull(x) else datetime(x.year + (x.month // 12), ((x.month % 12) + 1), 1) )
bd["mes"] = bd.next_f.apply(lambda x : x if pd.isnull(x) else int(x.month) )
bd["anio"] = bd.next_f.apply(lambda x : x if pd.isnull(x) else int(x.year) )
bd["corresp_date"] = bd.next_f.apply(lambda x : x if pd.isnull(x) else "{}-{:02}".format(x.year, x.month) )
bd.info()
```

```python
bd.sort_values(["anio", "mes"], inplace=True)
bd.head()
```

```python
columns = bd.corresp_date.unique()
row = bd.CUSTOMER_ID.unique()
```

```python
#cargo el dataset de cant por cliente, para mantener las mismas filas y columnas
bd["demora"] = 0
bd.loc[bd["DUE_DATE"] != bd["PAYMENT_DATE"], ["demora"]] = 1
bd.head()
```

```python
#ver dsp esto
#bd[bd["PAYMENT_STATUS"] == "AM"]
```

```python
bd["pago_deuda"] = 1
bd.loc[bd["PAYMENT_STATUS"] == "CT", "pago_deuda"]  = 0
```

```python
bd.head()
```

```python
bd.to_csv("../../data/processed/interm-BD_INT_PAGO_POLIZA_demora_pago.completo.csv")
```

# creando matriz demora pago

```python
df = pd.read_csv("../../data/processed/interm-BD_INT_PAGO_POLIZA_demora_pago.completo.csv")
df = df.iloc[:,1:]
df.info()
```

```python
df.head()
```

```python
columns = df.corresp_date.unique()
row = df.CUSTOMER_ID.unique()
```

```python
#creo el dataset vacio
df_m = pd.DataFrame(columns=columns)
df_m.info()
```

```python
#agrupo por cliente y mes_anio y cuento cuantas polizas distintas tiene
groups = df.groupby(["CUSTOMER_ID", "corresp_date"])["demora"].sum()
```

```python
for id_cliente in groups.index.levels[0]:
    df_m.loc[id_cliente] = groups[id_cliente].T
```

```python
df_m.info()
```

```python
df_m.head()
```

```python
df_m.to_csv("../../data/processed/BD_ID_CLIENTE_N.DEMORA_X_MES.csv")
```
