```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table("../../data/raw/TB_INTERACCIONES.tsv", encoding="latin-1")
df.info()
```

```python
from dateutil.parser import parse

df["fecha"] = df["INTERACTION_DATE"].apply(lambda x: x if pd.isnull(x) else x.split(" ")[0])
df["fecha"] = pd.to_datetime(df["fecha"], format="%d/%m/%Y")
```

```python
df["corresp_date"] = df.fecha.apply(lambda x : x if pd.isnull(x) else "{}-{:02}".format(x.year, x.month) )
df["mes"] = df.fecha.apply(lambda x : x if pd.isnull(x) else int(x.month) )
df["anio"] = df.fecha.apply(lambda x : x if pd.isnull(x) else int(x.year) )
```

```python
df.corresp_date.sort_values().value_counts()
```

# Versión simple contar cantidad

```python
df.head()
```

```python
df = df[df.corresp_date.notnull()]
```

```python
df.sort_values(["anio", "mes"], inplace=True)
```

```python
columns = df.corresp_date.unique()
row = df.CUSTOMER_ID.unique()
```

```python
#creo el dataset vacio
df_m = pd.DataFrame(columns=columns)
df_m.info()
```

```python
#agrupo por cliente y mes_anio y cuento cuantas interacciones distintas tiene
groups = df.groupby(["CUSTOMER_ID", "corresp_date"])["CUSTOMER_ID"].count()
groups
```

```python
for id_cliente in groups.index.levels[0]:
    df_m.loc[id_cliente] = groups[id_cliente].T
```

```python
df_m.to_csv("../../data/processed/4.4/tmp/BD_ID_CLIENTE_N.INTERACCIONES-GENERALES_X_MES.csv")
```
