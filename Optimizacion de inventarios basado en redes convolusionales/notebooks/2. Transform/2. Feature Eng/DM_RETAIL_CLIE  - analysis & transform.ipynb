# analysis

```python
# setting configs and constants
# path to dataset
PATH = "../../../data/interim/1. col_red"
FILENAME = "DM_RETAIL_CLIE-red_col.feather"
FILE = os.path.join(PATH, FILENAME)

if not(os.path.exists(FILE) and os.path.isfile(FILE)):
    raise Exception("File doesn't exists.")
```

## Loading database

```python
# usecols = []
df = pd.read_feather(FILE)
df.info()
```

## formatting

```python
df["ACA"] = df["ACA"] == "S"
```

```python
df["APTO_E_MAIL"] = df["APTO_E_MAIL"] == "S"
```

```python
df["FLAG_EBILLING"] = df["FLAG_EBILLING"] == "S"
```

```python
df["FLAG_ROBO_CONTENIDO"] = df["FLAG_ROBO_CONTENIDO"] == "S"
```

```python
df["MCA_0KM"] = df["MCA_0KM"].map({"N":0, "S":1})
```
