```python
import os
import pickle

import pandas as pd
```

```python
path = "../../../data/raw/CP_zonas.csv"

if not os.path.isfile(path):
    raise Exception("FIle not found")
```

```python
df = pd.read_csv(path, sep=";")
df.head()
```

# creating mapper for the 3 types of clusters

```python
cod_postal_mapper = {}
for col in ["DP_CLUSTER_2019", "RT_CLUSTER_2019", "RC_CLUSTER_2019"]:
    cod_postal_mapper[col] = dict(zip(df["COD_POSTAL"], df[col]))
cod_postal_mapper
```

# save it with pickle

```python
path = "../../../src/features/cod_postal_to_cluster_mapper.pickle"
with open(path, "wb") as handle:
    pickle.dump(cod_postal_mapper, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
