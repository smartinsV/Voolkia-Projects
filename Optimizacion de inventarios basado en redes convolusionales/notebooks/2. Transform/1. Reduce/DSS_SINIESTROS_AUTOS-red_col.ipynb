### Feature Format

```python
to_date = ['FECHA_SINI', 'FEC_DENU_SINI', 'FECHA_NAC_ASEG', 'FECHA_NAC_TERC', ]
date_format = "%d/%m/%Y"
# for is faster
# df.loc[:, to_date] = df.loc[:, to_date].apply(pd.to_datetime(format=date_format))
```

```python
%%time 
for col in to_date:
    df[col] = pd.to_datetime(df[col], format=date_format, errors="coerce")
```

# saving into feather file

```python
# fixing to_feather error
df["NUM_SECU_POL"] = df["NUM_SECU_POL"].astype("float64")
```

```python
to_drop = []
for v in df["NUM_SECU_POL"].unique():
    try:
        float(v)
    except Exception as e:
        print(f"{e}: {v}")
        to_drop.append(v)
to_drop
```

```python
df = df[~df["NUM_SECU_POL"].isin(to_drop)]
```
