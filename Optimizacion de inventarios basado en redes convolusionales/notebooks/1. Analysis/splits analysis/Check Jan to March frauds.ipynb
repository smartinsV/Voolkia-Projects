```python
import os

import pandas as pd
```

```python
# configs
pd.set_option("display.max_columns", 500)
pd.set_option("display.max_rows", 500)
```

## Loading Datasets

```python
# validation set (01-03 2019)

filename = "../../../data/split/1.1 - df_val 01-19to03-19.feather"

# check if file exists
if not os.path.isfile(filename):
    raise Exception(f"File {filename} doesn't exists")
```

```python
df = pd.read_feather(filename)
df.head()
```

```python
# validation info
df_val = pd.read_csv("fraudes eneAmar19.csv", sep=";")
df_val.info()
```

```python
df_val["FECHA_INI_INVEST"] = pd.to_datetime(df_val["FECHA_INI_INVEST"], format="%d/%m/%Y")
df_val.head()
```

## comparing number of frauds

```python
df["EXISTE_FRAUDE"].value_counts(dropna=False)
```

```python
len(df_val)
```

```python
len(df[df["EXISTE_FRAUDE"] == True]) - len(df_val)
```

**11 frauds of difference**

## getting the missing frauds

```python
our_frauds = set(df[df["EXISTE_FRAUDE"] == True]["NUM_SECU_EXPED"])
valid = set(df_val["NUM_SECU_EXPED"])

diff = valid - our_frauds
print(len(diff))
print(diff)
```

## Analyzing missing cases

```python
df_val[df_val["NUM_SECU_EXPED"].isin(diff)]
```

```python
df[df["NUM_SECU_EXPED"].isin(diff)][["NUM_SECU_EXPED", "EXISTE_FRAUDE"]]
```

```python
# there is 1 exped that we have as fraud but it's not.
our_frauds - valid
```
